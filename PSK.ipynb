```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
unipolar_arr = np.random.randint(0, 2, 10)
bipolar = 2 * unipolar_arr - 1 # convert input array to either -1 or 1
amp = 1
freq = 3 / amp  # carrier frequency
n_samples = 1000
time = np.linspace(0, 5, n_samples)

samples_per_bit = n_samples / unipolar_arr.size
dd = np.repeat(unipolar_arr, samples_per_bit) # normal wave
bb = np.repeat(bipolar, samples_per_bit) # reverse wave
waveform = amp * np.sin(2 * np.pi * freq * time)
bpsk_w = bb * waveform

f, ax = plt.subplots(2,1, sharex=True, sharey=True, squeeze=True)
ax[0].plot(time, bb)
ax[1].plot(time, bpsk_w, '.')
plt.grid(True)
ax[0].axis([0, 5, -1.5, 1.5])

plt.show()
```

```python
samples = 8 # samples

x = np.array([])
for bit in unipolar_arr:
    pulse = np.zeros(samples)
    pulse[0] = bit*2-1 # set the first value to either a 1 or -1
    x = np.concatenate((x, pulse)) # add the 8 samples to the signal
    
plt.plot(x, '.-')
plt.grid(True)
plt.show()
```
